```python
train = get_train_data(cutoff_date = None)
print(train)
```

```python
most_populer_items = list((train['article_id'].value_counts()).index)[:12]
print(most_populer_items)
```

```python
recent_purchases = train.loc[train.t_dat >= pd.to_datetime('2020-08-23')]
recent_purchases['customer_id'].unique().shape
```

```python
frequency_list = {}
for idx, row in recent_purchases.iterrows():
    customer_id, article_id = row['customer_id'], row['article_id']
    
    if customer_id not in frequency_list.keys():
        frequency_list[customer_id] = []
    frequency_list[customer_id].append(article_id)
```

```python
for customer_id in list(frequency_list.keys()):
    print(frequency_list[customer_id])
    frequency_list[customer_id] = list((pd.DataFrame(frequency_list[customer_id])[0].value_counts()).index)
    print(frequency_list[customer_id])
```

```python
submission = load_csv('sample_submission.csv')
print(submission[['customer_id']])
```

```python
predictions = []
for customer_id in submission['customer_id']:
    if customer_id in frequency_list.keys():
        if len(frequency_list[customer_id])>=12:
            prediction = " ".join(frequency_list[customer_id][:12])
        else:
            prediction = " ".join((frequency_list[customer_id] + most_populer_items)[:12])
    else:
        prediction = " ".join(most_populer_items)
    predictions.append(prediction)
submission['prediction'] = predictions
print(submission)
```
